```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.zeros((1, 5)))
df.columns = ['status', 'title', 'text', 'date', 'keywords']
```

Примеры: [Парсер для WordPress](https://github.com/pavkozlov/parsing/blob/master/wordpress/parser.py)

# Habr

```python
BASE_URL = 'https://habr.com/ru/post/'
```

```python
def parser(pid, print_each=2):
    global df
    r = requests.get('https://habr.com/ru/post/'+ str(pid)+'/') # парсинг сайта
    if r.status_code == 200:
        soup = BeautifulSoup(r.text, 'html5lib')
    else:
        return None
    doc = {}
    # if not soup.find('h1', {'class': 'tm-article-snippet__title_h1'}): # если статья не существовала или удалена
    #     doc['status'] = 'title_not_found' 
    # else:
    try:
        doc['status'] = 'ok'
        doc['title'] = soup.find('h1', {'class': 'tm-article-snippet__title_h1'}).text
        dir = soup.find('div', {'class': 'article-formatted-body'})
        doc['text'] = dir.find('div').text
        doc['date'] = soup.find('div', {'class': 'tm-article-snippet__meta-container'}).find('span', {'class': 'tm-article-snippet__datetime-published'}).text
        tags = soup.find_all('span', {'class': 'tm-article-snippet__hubs-item'})
        all_topics = []
        for tag in tags:
            topic = tag.find('span').text
            all_topics.append(topic)
        doc['keywords'] = all_topics
    except AttributeError:
        pass

    df = df.append(doc, ignore_index=True)
    print(doc['status'])
```

```python
for pid in range(345000, 350000):
    parser(pid)
```
